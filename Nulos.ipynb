```python
#2. Selección de Método de Imputación: Discute en tu compañera cuál sería la mejor estrategia para manejar los valores nulos en cada una de las columnas identificadas en el paso anterior.

#minimum_wage -->45: Es la columna que tiene más nulos, pero vamos a hacer una mediana para encontrar el punto medio entre los países que tienen un mínimo más elevado y los que tienen un mínimo más bajo. 


df["minimum_wage"].median()

#La media es 1.045. Tenemos que pensar si lo imputamos a todos los nulos de este df. 
```

```python
# la primera columna que gestionaremos sera 'conspriceidx' usando el método 'fillna()' que ya conocemos
# calculamos la media para la columna 'conspriceidx'
media_minimum_wage = df["minimum_wage"].median()
print(f"La media de la columna 'minimum_wage' es: {round(media_minimum_wage, 2)}")

# aplicamos el método 'fillna()' a la columna
df["minimum_wage"] = df["minimum_wage"].fillna(media_minimum_wage)

# comprobamos los nulos para la columna
print(f"Después del 'fillna' tenemos {df['minimum_wage'].isnull().sum()} nulos")
```

```python
# Hemos decidido hacer un KNN en el tamaño de la armada ya que dependiendo de los países vecinos 
# tendremos más tendencia a tener armada y a que ésta tenga un tamaño en concreto. 

from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Removemos comas en la columna 'armed_forces_size' y convertimos a numérico
df["armed_forces_size"] = df["armed_forces_size"].replace({',': ''}, regex=True)
df["armed_forces_size"] = pd.to_numeric(df["armed_forces_size"], errors='coerce')

# Creamos una copia del DataFrame solo con las columnas necesarias
df_impute = df[["armed_forces_size", "country"]].copy()

# Convertimos 'country' a valores numéricos si es categórico
if df_impute["country"].dtype == object:
    df_impute["country"] = LabelEncoder().fit_transform(df_impute["country"])

# Instanciamos el imputador KNN
imputer_knn = KNNImputer(n_neighbors=5)

# Aplicamos el imputador KNN
imputer_knn_imputado = imputer_knn.fit_transform(df_impute)

# Convertimos el resultado a un DataFrame
df_imputed = pd.DataFrame(imputer_knn_imputado, columns=["armed_forces_size", "country"])

# Actualizamos la columna 'armed_forces_size' en el DataFrame original
df["armed_forces_size"] = df_imputed["armed_forces_size"]

# Comprobamos el resultado
print(df["armed_forces_size"])
```
